1. What is hypothesis testing in statistics?

Ans- Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It involves two hypotheses:
- Null Hypothesis (H‚ÇÄ): Assumes no effect or difference.
- Alternative Hypothesis (H‚ÇÅ): Assumes there is an effect or difference.

The process includes selecting a significance level (commonly 0.05), calculating a test statistic (like a z or t-value), and determining a p-value. If the p-value is less than or equal to the significance level, we reject H‚ÇÄ; otherwise, we fail to reject it.

This method helps evaluate claims and make decisions using data, while managing risks of Type I and Type II errors.



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Ans- The null hypothesis (H‚ÇÄ) is a statement used in statistics that assumes there is no effect, no difference, or no relationship between variables. It represents the default or status quo condition.

The alternative hypothesis (H‚ÇÅ or Ha) is the statement that contradicts the null hypothesis. It suggests that there is an effect, a difference, or a relationship.
- Null Hypothesis (H‚ÇÄ)       
 - Meaning: Assumes no change or effect
 - Purpose: 	Acts as a baseline to test against
 - Decision Rule: 	Reject or fail to reject
- Alternative Hypothesis (H‚ÇÅ)
 - Meaning: Assumes there is a change or effect
 - Purpose: 	Represents what the researcher wants to prove
 - Decision Rule: Never directly accepted, only supported if H‚ÇÄ is rejected

3. What is the significance level in hypothesis testing, and why is it important?

Ans- The significance level (denoted by Œ±) is the threshold used in hypothesis testing to decide whether to reject the null hypothesis (H‚ÇÄ). It represents the probability of making a Type I error, which is rejecting H‚ÇÄ when it is actually true.

Why It‚Äôs Important:
- Controls Error Risk:
 - It sets the maximum chance you're willing to take of incorrectly rejecting a true null hypothesis.
- Decision-Making:
 - If the p-value ‚â§ Œ±, you reject H‚ÇÄ (result is statistically significant).
 - If p-value > Œ±, you fail to reject H‚ÇÄ.
- Scientific Rigor:
 - A smaller Œ± (like 0.01) demands stronger evidence to reject H‚ÇÄ, increasing confidence in results.

14. > What is a confidence interval, and how is it used to interpret statistical results?

Ans- A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate a population parameter (such as the population mean or population proportion). It provides an interval estimate of the parameter, along with a confidence level that indicates the probability that the interval contains the true population parameter.

A confidence interval is not a guarantee that the population parameter is within the interval, but it gives us a range that is likely to contain the true parameter.
- For example, a 95% confidence interval of (45, 55) for the population mean means we are 95% confident that the true population mean lies between 45 and 55.
- 95% confidence means that if we were to take 100 different samples and construct 100 different confidence intervals, approximately 95 of those intervals would contain the true population parameter.

15. What is the margin of error, and how does it affect the confidence interval?

Ans- The margin of error (MOE) is a statistical term that quantifies the uncertainty or variability in the estimate of a population parameter based on sample data. It provides the range within which the true population parameter is likely to fall. The margin of error is a critical component of the confidence interval (CI).

The margin of error directly influences the width of the confidence interval.
- Larger Margin of Error = Wider Confidence Interval:
A larger margin of error leads to a wider confidence interval, meaning there's greater uncertainty about where the true population parameter lies.
- Smaller Margin of Error = Narrower Confidence Interval:
A smaller margin of error leads to a narrower confidence interval, implying a more precise estimate of the population parameter.

16. How is Bayes' Theorem used in statistics, and what is its significance?

Ans- Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update the probability of an event, given new evidence. It is named after the Reverend Thomas Bayes, who introduced it in the 18th century. Bayes' Theorem provides a method to revise prior beliefs or probabilities in light of new data, and it is a core concept in Bayesian statistics.

Significance of Bayes' Theorem:
- Updating Beliefs:

Bayes' Theorem allows for continuous updating of knowledge based on new data. This is particularly useful in real-time situations, where data is constantly being collected, and previous beliefs or estimates need to be adjusted accordingly.
- Incorporating Prior Knowledge:

One of the main advantages of Bayes' Theorem is that it allows us to incorporate prior knowledge or expert judgment (prior probability) into the analysis, which is often very valuable in real-world scenarios where direct data may be limited.
- Dealing with Uncertainty:

Bayesian methods are especially powerful for handling uncertainty and making probabilistic predictions. The posterior probability is a reflection of how likely an event is, given all available evidence, which is more informative than a simple binary yes/no decision.
- Flexibility in Modeling:

Bayes' Theorem is flexible and can be used in a wide variety of statistical models, from simple problems (like coin flipping) to complex ones (like hierarchical models or machine learning algorithms). Its ability to incorporate various forms of data and prior information makes it applicable to a broad range of fields.
- Bayesian vs. Frequentist Methods:

While frequentist statistics focus on the likelihood of data under a fixed model, Bayesian statistics focus on the probability of hypotheses or parameters, allowing for more intuitive interpretations and updating beliefs. This philosophical difference makes Bayesian methods particularly suitable for certain types of inference, especially when prior information is important.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low = 1      # Inclusive lower bound
high = 7     # Exclusive upper bound ‚Üí values: 1 to 6

# Create the discrete uniform distribution
dist = randint(low, high)

# Define the support (possible values)
x = np.arange(low, high)

# Get PMF values for each x
pmf_values = dist.pmf(x)

# Plotting the PMF
plt.figure(figsize=(8, 4))
# Remove the 'use_line_collection' argument
plt.stem(x, pmf_values)
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution>

```python
def bernoulli_pdf(x, p):
    """
    Calculates the PDF (PMF) of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1)
        p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
        float: Probability of observing x
    """
    if x == 1:
        return p
    elif x == 0:
        return 1 - p
    else:
        return 0  # Invalid input for a Bernoulli variable
```

```python
# Probability of success
p = 0.7

# Calculate probability for outcome 1 (success)
print("P(X=1):", bernoulli_pdf(1, p))

# Calculate probability for outcome 0 (failure)
print("P(X=0):", bernoulli_pdf(0, p))

# Invalid outcome (e.g., 2)
print("P(X=2):", bernoulli_pdf(2, p))  # Returns 0
```

9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores from a dataset and plots them.

    Parameters:
        data (list or np.array): The input dataset.

    Returns:
        np.array: Array of Z-scores.
    """
    # Convert to NumPy array
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Avoid division by zero
    if std == 0:
        raise ValueError("Standard deviation is zero; Z-scores cannot be computed.")

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot Z-scores
    plt.figure(figsize=(8, 4))
    plt.plot(z_scores, 'bo-', markersize=4, label='Z-scores')
    plt.axhline(0, color='gray', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-scores of Dataset')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True)
    plt.legend()
    plt.show()

    return z_scores
```

```python
# Sample data
data = [10, 12, 13, 15, 18, 20, 22, 25, 30]

# Call the function
z_scores = calculate_and_plot_z_scores(data)

# Print Z-scores
print("Z-scores:", z_scores)
```

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for the exponential distribution (non-normal)
lambda_val = 1.0
population_size = 100000
population = np.random.exponential(scale=1/lambda_val, size=population_size)

# CLT simulation parameters
sample_size = 30     # Size of each sample
num_samples = 1000   # Number of samples

# Collect means of the samples
sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=False)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Plotting the results
plt.figure(figsize=(10, 5))

# Histogram of original population (non-normal)
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='orange', alpha=0.7, edgecolor='black')
plt.title('Original Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram of sample means (should approximate normal)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, color='skyblue', alpha=0.7, edgecolor='black')
plt.title(f'Distribution of Sample Means (n={sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for the sample data.

    Parameters:
        sample_data (list or np.array): The sample data.
        population_mean (float): The mean of the population (null hypothesis).
        population_std (float): The standard deviation of the population.
        alpha (float): Significance level.

    Returns:
        None
    """
    # Calculate the sample mean and size
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed p-value

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Make decision based on p-value
    if p_value < alpha:
        print(f"Reject the null hypothesis (p-value < {alpha}).")
    else:
        print(f"Do not reject the null hypothesis (p-value >= {alpha}).")

# Example data: Sample dataset
sample_data = [115, 120, 125, 130, 125, 110, 105, 115, 120, 130]

# Parameters for the hypothesis test
population_mean = 120  # Null hypothesis: The population mean is 120
population_std = 10    # Assume the population standard deviation is known and is 10
alpha = 0.05           # Significance level

# Perform the Z-test
z_test(sample_data, population_mean, population_std, alpha)
```

16. Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for the mean of a sample dataset.

    Parameters:
        data (list or np.array): The sample data.
        confidence_level (float): The desired confidence level (default is 0.95 for 95%).

    Returns:
        tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample mean
    sample_mean = np.mean(data)

    # Sample standard deviation
    sample_std = np.std(data, ddof=1)  # Using ddof=1 for sample standard deviation

    # Sample size
    n = len(data)

    # Z-value for the desired confidence level (95% confidence level uses Z=1.96)
    z_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)

    # Margin of error
    margin_of_error = z_value * (sample_std / np.sqrt(n))

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example dataset: Sample data
data = [120, 125, 130, 110, 115, 135, 125, 140, 120, 110]

# Calculate 95% confidence interval for the mean
lower, upper = confidence_interval(data, confidence_level=0.95)

# Print the result
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type_errors(population_mean, alternative_mean, population_std, alpha=0.05, sample_size=30):
    """
    Visualizes Type 1 and Type 2 errors during hypothesis testing.

    Parameters:
    - population_mean: Mean under the null hypothesis
    - alternative_mean: Mean under the alternative hypothesis
    - population_std: Known population standard deviation
    - alpha: Significance level (default 0.05)
    - sample_size: Sample size (default 30)
    """
    # Define the critical z-value for a two-tailed test at alpha significance level
    z_critical = norm.ppf(1 - alpha / 2)

    # Define the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Z-score corresponding to the critical value
    critical_value = population_mean + z_critical * standard_error

    # Plotting the normal distributions for the null hypothesis (H0) and alternative hypothesis (H1)
    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)
    y_null = norm.pdf(x, population_mean, population_std / np.sqrt(sample_size))
    y_alt = norm.pdf(x, alternative_mean, population_std / np.sqrt(sample_size))

    plt.figure(figsize=(10, 6))

    # Plot the null hypothesis (H0) and alternative hypothesis (H1) distributions
    plt.plot(x, y_null, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, y_alt, label="Alternative Hypothesis (H1)", color="orange")

    # Shade the Type 1 error region (rejection region under H0)
    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.5, label="Type 1 Error (Rejection Region)")

    # Shade the Type 2 error region (failing to reject under H1)
    plt.fill_between(x, y_alt, where=(x <= critical_value), color='green', alpha=0.5, label="Type 2 Error (Non-rejection Region)")

    # Mark the critical value (threshold for rejection)
    plt.axvline(critical_value, color="black", linestyle="--", label=f"Critical Value: {critical_value:.2f}")

    # Title and labels
    plt.title("Visualization of Type 1 and Type 2 Errors")
    plt.xlabel("Test Statistic (Sample Mean)")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper left")
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage:
population_mean = 100  # Null hypothesis mean
alternative_mean = 105  # Alternative hypothesis mean (if H0 is false)
population_std = 15  # Known population standard deviation

# Visualize Type 1 and Type 2 errors
visualize_type_errors(population_mean, alternative_mean, population_std)
```

6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
from scipy import stats

def independent_t_test(group1, group2):
    """
    Perform an independent t-test to compare the means of two groups.

    Parameters:
    - group1: List or array of data for group 1
    - group2: List or array of data for group 2

    Returns:
    - t_statistic: Calculated t-statistic
    - p_value: Calculated p-value
    """
    # Calculate the t-statistic and p-value using scipy.stats.ttest_ind
    t_stat, p_value = stats.ttest_ind(group1, group2)

    # Print the results
    print(f"Group 1 Mean: {np.mean(group1):.2f}, Standard Deviation: {np.std(group1, ddof=1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}, Standard Deviation: {np.std(group2, ddof=1):.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the two groups.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the two groups.")

    return t_stat, p_value

# Example usage:
np.random.seed(42)  # For reproducibility

# Simulate two independent samples (e.g., test scores for two different teaching methods)
group1 = np.random.normal(loc=75, scale=10, size=30)  # Group 1 (mean = 75, std = 10)
group2 = np.random.normal(loc=80, scale=10, size=30)  # Group 2 (mean = 80, std = 10)

# Perform the Independent T-test
independent_t_test(group1, group2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_t_test(before, after):
    """
    Perform a paired sample t-test to compare the means of two related groups.

    Parameters:
    - before: List or array of data for the 'before' condition
    - after: List or array of data for the 'after' condition

    Returns:
    - t_statistic: Calculated t-statistic
    - p_value: Calculated p-value
    """
    # Perform paired T-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Print results
    print(f"Before Mean: {np.mean(before):.2f}, Standard Deviation: {np.std(before, ddof=1):.2f}")
    print(f"After Mean: {np.mean(after):.2f}, Standard Deviation: {np.std(after, ddof=1):.2f}")
    print(f"T-Statistic: {t_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")

    # Interpretation
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference between the before and after groups.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference between the before and after groups.")

    return t_stat, p_value

def visualize_comparison(before, after):
    """
    Visualize the comparison between the before and after values.
    """
    # Create a list of indices to plot
    indices = np.arange(len(before))

    # Plot before vs after data
    plt.figure(figsize=(10, 6))
    plt.plot(indices, before, label="Before", marker='o', linestyle='-', color='blue')
    plt.plot(indices, after, label="After", marker='o', linestyle='-', color='orange')

    # Add a line to show the differences
    for i in range(len(before)):
        plt.plot([indices[i], indices[i]], [before[i], after[i]], color='gray', linestyle='--', alpha=0.5)

    # Add labels and title
    plt.title("Paired Sample T-Test Comparison (Before vs After)")
    plt.xlabel("Sample Index")
    plt.ylabel("Measurement Value")
    plt.legend()
    plt.grid(True)

    # Show the plot
    plt.show()

# Example usage:
np.random.seed(42)  # For reproducibility

# Simulate 'before' and 'after' treatment data
before = np.random.normal(loc=100, scale=15, size=30)  # Before treatment (mean = 100, std = 15)
after = before + np.random.normal(loc=5, scale=10, size=30)  # After treatment (mean increase of ~5)

# Perform the Paired T-test
paired_t_test(before, after)

# Visualize the comparison between the before and after values
visualize_comparison(before, after)
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate data for a sample
np.random.seed(42)  # For reproducibility

population_mean = 100
population_std = 15
sample_size = 30

# Generate a random sample from a normal distribution
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Z-test (assuming population std is known)
def z_test(sample, population_mean, population_std):
    sample_mean = np.mean(sample)
    standard_error = population_std / np.sqrt(len(sample))
    z_stat = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_stat)))  # Two-tailed p-value
    return z_stat, p_value

# T-test (assuming population std is unknown)
def t_test(sample, population_mean):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(len(sample))
    t_stat = (sample_mean - population_mean) / standard_error
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df=len(sample)-1))  # Two-tailed p-value
    return t_stat, p_value

# Perform Z-test and T-test
z_stat, z_p_value = z_test(sample, population_mean, population_std)
t_stat, t_p_value = t_test(sample, population_mean)

# Print results
print(f"Z-Test: Z-Statistic = {z_stat:.4f}, P-Value = {z_p_value:.4f}")
print(f"T-Test: T-Statistic = {t_stat:.4f}, P-Value = {t_p_value:.4f}")

# Interpret results
alpha = 0.05
print("\nInterpretation:")
if z_p_value < alpha:
    print("Z-test: Reject the null hypothesis (significant difference).")
else:
    print("Z-test: Fail to reject the null hypothesis (no significant difference).")

if t_p_value < alpha:
    print("T-test: Reject the null hypothesis (significant difference).")
else:
    print("T-test: Fail to reject the null hypothesis (no significant difference).")

# Visualize the distribution of the sample
plt.figure(figsize=(10, 6))
x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)
plt.plot(x, stats.norm.pdf(x, population_mean, population_std), label="Population Distribution", color="blue", linestyle="--")
plt.hist(sample, bins=10, density=True, alpha=0.7, color="green", label="Sample Distribution")

plt.title("Comparison of Population and Sample Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def confidence_interval(sample, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - sample: List or array of sample data
    - confidence_level: Desired confidence level (default is 0.95)

    Returns:
    - lower_bound: Lower bound of the confidence interval
    - upper_bound: Upper bound of the confidence interval
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Use sample standard deviation (ddof=1)
    n = len(sample)

    # Find the critical value (z or t) depending on confidence level and sample size
    if n > 30:  # Z-test approximation when sample size is large
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)  # z-score for normal distribution
    else:  # t-distribution for small sample sizes
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=n-1)  # t-score for t-distribution

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(n)

    # Margin of error
    margin_of_error = critical_value * standard_error

    # Confidence interval bounds
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)  # For reproducibility

# Simulate a sample of data
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Mean=100, SD=15, n=30

# Calculate 95% confidence interval
lower, upper = confidence_interval(sample_data, confidence_level=0.95)

print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
# Bayesian Inference: Calculate Posterior Probability using Bayes' Theorem

def bayesian_inference(prior_H, likelihood_E_given_H, likelihood_E_given_Hc, prior_Hc):
    """
    Apply Bayes' Theorem to calculate the posterior probability of a hypothesis (H) given evidence (E).

    Parameters:
    - prior_H: Prior probability of hypothesis H (e.g., P(H))
    - likelihood_E_given_H: Likelihood of evidence given H (e.g., P(E|H))
    - likelihood_E_given_Hc: Likelihood of evidence given not H (e.g., P(E|H^c))
    - prior_Hc: Prior probability of not H (e.g., P(H^c))

    Returns:
    - posterior_H_given_E: Posterior probability of hypothesis H given evidence E (P(H|E))
    """
    # Calculate the evidence (P(E)) = P(E|H) * P(H) + P(E|H^c) * P(H^c)
    evidence = (likelihood_E_given_H * prior_H) + (likelihood_E_given_Hc * prior_Hc)

    # Calculate the posterior probability using Bayes' Theorem
    posterior_H_given_E = (likelihood_E_given_H * prior_H) / evidence

    return posterior_H_given_E

# Example values
prior_H = 0.01  # P(H) = 1% chance the patient has the disease
likelihood_E_given_H = 0.9  # P(E|H) = 90% chance the test is positive given the patient has the disease
likelihood_E_given_Hc = 0.05  # P(E|H^c) = 5% chance of a false positive
prior_Hc = 0.99  # P(H^c) = 99% chance the patient does not have the disease

# Calculate the posterior probability P(H|E)
posterior_H_given_E = bayesian_inference(prior_H, likelihood_E_given_H, likelihood_E_given_Hc, prior_Hc)

# Output the result
print(f"The posterior probability that the patient has the disease given a positive test result is: {posterior_H_given_E:.4f}")
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate the expected frequencies for a Chi-Square test based on the observed data.

    Parameters:
    - observed: A 2D NumPy array representing the observed contingency table.

    Returns:
    - expected: A 2D NumPy array with the expected frequencies.
    """
    # Get the row totals and column totals
    row_totals = observed.sum(axis=1, keepdims=True)
    column_totals = observed.sum(axis=0, keepdims=True)

    # Calculate the total sum of all observed values
    total_sum = observed.sum()

    # Calculate the expected frequencies using the formula: E = (row total * column total) / grand total
    expected = (row_totals * column_totals) / total_sum

    return expected

# Example observed data (Contingency table)
# For example, the table could represent gender (Male, Female) and choice of beverage (Tea, Coffee)
observed_data = np.array([[30, 10],  # 30 males prefer tea, 10 males prefer coffee
                          [20, 40]])  # 20 females prefer tea, 40 females prefer coffee

# Calculate the expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_data)

# Output the observed and expected frequencies
print("Observed Frequencies:\n", observed_data)
print("\nExpected Frequencies:\n", expected_frequencies)
```
